```python
import matplotlib.pyplot as plt 
import numpy as np
import cv2
```

```python
img = cv2.imread("Apple.jpg")
# kernal= np.array([[-1,-2,-1],
#                   [0,0,0],
#                   [1,2,1]])
sobal_img = cv2.Sobel(img ,-1,0,1,1)
canny_img = cv2.Canny(img , 300 ,300 )
```

```python
fig = plt.figure(figsize=(15,15))
title = ["Original image ", "Sobal Filter" , "Canny Filter"]
image = img , sobal_img , canny_img

for i in range (3):
    plt.subplot(2,3,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
plt.show
```
