```python
import matplotlib.pyplot as plt 
import numpy as np
import cv2
```

Gaussion Blur

```python
img  = cv2.imread('baboon.jpg')
```

```python
kernal_size = (5,5)
sigma = 1

Gaussion_img = cv2.GaussianBlur(img,kernal_size , sigma)
```

Bilateral Blur 

```python
bilateral_img = cv2.bilateralFilter(img ,9,75,75 )
```

SHARPENING IMAGE

```python
kernal= np.array([[-1,-1,-1],
          [-1,9,-1],
          [-1,-1,-1]])
sharp_img = cv2.filter2D(img , -1 ,kernal)
```

PLOTING 

```python
title = ["Original_Image","Gaussion blur" , "Bilateral_blur " , "Sharpening image"]
images = img , Gaussion_img , bilateral_img , sharp_img
```

```python
fig = plt.figure(figsize=(15,15))
for i in range (4):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i])
    plt.title(title[i])
plt.show
```
