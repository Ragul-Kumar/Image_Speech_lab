```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt

# Load audio files with librosa
d_file = "Audio/bed.wav"
d1_file = "Audio/cat.wav"
d, sr = librosa.load(d_file, sr=8000)
d1, sr = librosa.load(d1_file, sr=8000)

# Plot the first audio signal
plt.subplot(2, 2, 1)
t = np.linspace(0, len(d) / sr, len(d))
plt.plot(t, d, lw=1)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal 1')

# Plot the second audio signal
plt.subplot(2, 2, 2)
t = np.linspace(0, len(d1) / sr, len(d1))
plt.plot(t, d1, lw=1)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal 2')

# Convolve the two audio signals
y = signal.convolve(d, d1, mode='full')

# Plot the convolved signal
plt.subplot(2, 2, 3)
t = np.linspace(0, len(y) / sr, len(y))
plt.plot(t, y, lw=1)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Convolution of Signals')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
